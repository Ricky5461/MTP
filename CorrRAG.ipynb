```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1. BUILD THE ARCHITECTURE
class SimpleMLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleMLP, self).__init__()
        
        # Layer 1: The Hidden Layer (Connects Input -> Hidden)
        self.hidden_layer = nn.Linear(input_size, hidden_size)
        
        # The Non-Linear Activation Function (The secret sauce!)
        self.relu = nn.ReLU()
        
        # Layer 2: The Output Layer (Connects Hidden -> Output)
        self.output_layer = nn.Linear(hidden_size, output_size)

    # This defines how data flows FORWARD through the network
    def forward(self, x):
        x = self.hidden_layer(x)  # Math: Weights * Input + Bias
        x = self.relu(x)          # Bend the data (Non-linearity)
        x = self.output_layer(x)  # Final prediction
        return x

# --- SETTING UP THE AI ---

# Let's say we are predicting a house price based on 3 features (Bedrooms, Bathrooms, Age)
INPUT_FEATURES = 3
HIDDEN_NEURONS = 5
OUTPUT_PREDICTION = 1

# Initialize our model
model = SimpleMLP(INPUT_FEATURES, HIDDEN_NEURONS, OUTPUT_PREDICTION)

# Define the Loss Function (How we calculate the error)
# MSE (Mean Squared Error) is standard for predicting numbers
criterion = nn.MSELoss()

# Define the Optimizer (The Gradient Descent engine that updates the weights)
# Learning Rate (lr) is the size of the steps it takes to fix its mistakes
optimizer = optim.SGD(model.parameters(), lr=0.01)

# --- THE TRAINING LOOP ---

# Dummy Data: 1 house with 3 bedrooms, 2 baths, 10 years old
dummy_input = torch.tensor([[3.0, 2.0, 10.0]]) 
# The actual true price of the house (e.g., $300k)
true_answer = torch.tensor([[300.0]]) 

print("Starting Training...")

for epoch in range(5): # We will train it 5 times (epochs)
    
    # Step 1: Forward Pass (Make a guess)
    prediction = model(dummy_input)
    
    # Step 2: Calculate the Loss (How wrong was the guess?)
    loss = criterion(prediction, true_answer)
    
    # Step 3: Backpropagation (Calculate the gradients)
    optimizer.zero_grad() # Clear old gradients
    loss.backward()       # Figure out which weights caused the error
    
    # Step 4: Gradient Descent (Update the weights)
    optimizer.step()      
    
    print(f"Epoch {epoch+1} | AI's Guess: {prediction.item():.2f} | Loss (Error): {loss.item():.2f}")

print("Training Complete! Notice how the AI's guess gets closer to 300 each time.")
```
